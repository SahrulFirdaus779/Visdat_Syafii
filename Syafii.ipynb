# ETL

```python
df['Row ID'] = df['Row ID'].astype(str)
df['Postal Code'] = df['Postal Code'].astype(str)
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'],format='%m/%d/%Y',errors='coerce')
df['Ship Date']=pd.to_datetime(df['Ship Date'],format='%m/%d/%Y',errors='coerce')
```

```python
df['Delivery Time']=df['Ship Date']-df['Order Date']
df['Delivery Time']=df['Delivery Time'].astype('str')
df['Delivery Time']=df['Delivery Time'].apply(lambda x:x.split(' ')[0])
df['Delivery Time']=df['Delivery Time'].apply(lambda x:int(x))
df['Delivery Time']
```

```python
df['Order_Month']=df['Order Date'].dt.month_name()
df['Order_Day']=df['Order Date'].dt.day_name()
```

```python
df['Order_Year']  = df['Order Date'].dt.year
```

```python
# Menghitung persentase keuntungan dari setiap penjualan.
# Menggunakan .apply() dengan lambda untuk menangani kasus Sales = 0 dan menghindari pembagian oleh nol.
df['Profit_Ratio'] = df.apply(lambda row: row['Profit'] / row['Sales'] if row['Sales'] != 0 else 0, axis=1)
# Menghitung keuntungan yang dihasilkan per unit produk yang terjual.
# Menggunakan .apply() dengan lambda untuk menangani kasus Quantity = 0 dan menghindari pembagian oleh nol.
df['Profit_Per_Quantity'] = df.apply(lambda row: row['Profit'] / row['Quantity'] if row['Quantity'] != 0 else 0, axis=1)
```

## 3. Cek Outlier & Distribusi Variabel
Distribusi & boxplot untuk variabel penting

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style untuk visualisasi
sns.set(style='whitegrid')

# Kolom-kolom yang akan divisualisasikan
numeric_columns = ['Sales', 'Discount', 'Profit', 'Profit_Ratio', 'Profit_Per_Quantity']

# Buat grid plot: 2 baris × 5 kolom (histogram dan boxplot berdampingan)
plt.figure(figsize=(18, 12))

for i, col in enumerate(numeric_columns):
    # Histogram + KDE
    plt.subplot(5, 2, 2*i + 1)
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f'Distribusi {col}')
    plt.xlabel(col)

    # Boxplot
    plt.subplot(5, 2, 2*i + 2)
    sns.boxplot(x=df[col], color='lightcoral')
    plt.title(f'Boxplot {col}')
    plt.xlabel(col)

plt.tight_layout()
plt.show()
```
